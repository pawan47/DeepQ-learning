```python
import random
import gym
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import np_utils
from keras.layers import Flatten
from collections import deque

```

```python

memory = deque(maxlen=2000)
env=gym.make('CartPole-v0')
n=env.action_space.n
```

```python
model=Sequential()
model.add(Dense(12,input_shape=(1,4),activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(n,activation='relu'))
model.compile(loss='mse',optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
ep=1.00
ep_decay=0.9
gamma=0.95
lr=.1
num_episode=1000
batch_size=32
```

```python
env=gym.make('CartPole-v0')
env.reset()
env.render()
```

```python

for episode in range(num_episode):
    d=False
    R,r=0,0
    s=env.reset()
    state_size = env.observation_space.shape[0]
    #print(state_size)
    s=np.reshape(s,[1,1,state_size])

    cc=0
    while cc<500:
        cc+=1
        if np.random.random() < ep:
            a=np.random.randint(0,high=2)
            ep=ep*ep_decay
        else:

            a=model.predict(s)
            a=np.argmax(a[0])
        s2,r,d,_=env.step(a)
        if d:
            r=-10
        R=R+r
        s2=np.reshape(s2,[1,1,state_size])
        memory.append([s,a,r,s2,d])
        s=s2
        if d:
            break
    if len(memory) > batch_size:
        ll=random.sample(memory,batch_size)
        for s,a,r,s2,d in ll:
            t=r
            if not d:
                t=r+gamma*np.amax(model.predict(s2))
            target=model.predict(s)
            #print(target.shape)
            target[0][0][a]=t
            model.fit(s,target,batch_size=32,epochs=1,verbose=0)
    print ('num_episode {} total reward {}'.format(episode,cc))
```
